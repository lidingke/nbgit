```python
class Solution:
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if len(nums)<1:
            return 0
        left = 0
        last = nums[0]
        for i in range(1,len(nums)):
            if nums[i] != last:
                last = nums[i]

                left+=1
                if i != left:
                    nums[left],nums[i]=nums[i],nums[left]
        print(nums)
        return left+1
```

```python
class Solution:
    def findKthLargest(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        self.nums =nums
        def partition(i,j):
            root,r = self.nums[i],i
            for v in range(i,j):
                if self.nums[v]<root:
                    r+=1
#                     if v!=r:
                    self.nums[r],self.nums[v]=self.nums[v],self.nums[r]
                        
            self.nums[r],self.nums[i]=self.nums[i],self.nums[r]
#             print(self.nums,i,j,r)
                    
            return r
        k = len(nums)-k
        def find(i,j):
            print('k',k)
            
            mid = partition(i,j)
            if mid == k:
                print('==')
                return mid
            elif mid > k:
                mid = find(i,mid)
            else:
                mid = find(mid+1,j)
            return mid
                
        return self.nums[find(0,len(nums))]
    
    
```
